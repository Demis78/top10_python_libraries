```python
import scipy.fftpack
import numpy
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,6)
```

## create a signal of 2 waves combined

```python
x = numpy.linspace(0.0, 1.0, 500)
y1 = numpy.sin(10*numpy.pi*x)
y2 = numpy.sin(50*numpy.pi*x) * 0.5
```

```python
# visualise y1 and y2
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# combine y1 and y2
y = y1+ y2
plt.plot(x,y)
plt.show()
```

## reverse using FFT

```python
# fft from scipy

xf = numpy.linspace(0.0, 500 , 250)
# perform fft
yf = scipy.fftpack.fft(y)
# scaling gumpf
yf = 2 / 500 * numpy.abs(yf)
```

```python
# plot fft
plt.plot(xf, yf[0:250])
plt.grid()
plt.show()
```
