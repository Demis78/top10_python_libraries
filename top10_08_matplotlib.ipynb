```python
import matplotlib.pyplot as plt
import numpy
# set default size for figures
plt.rcParams['figure.figsize'] = (8,6)
%matplotlib inline
```

## simplest example

just a list of numbers

```python
A = numpy.array([1.0, 3.0, 2.0, 1.5])
print(A)
plt.plot(A)
plt.show()
```

## still very simple

x and y coordinates

```python
X = numpy.linspace(0, 10, 200)
Y = numpy.sin(X)
plt.plot(X,Y)
plt.show()
```

## histogram

use previous Y values

```python
plt.hist(Y, bins=50)
plt.show()
```

## 2D bitmap

plot a 2-d array of numbers

```python
data = numpy.zeros((100, 100))

# calculate some values
for x in range(data.shape[0]):
    for y in range(data.shape[1]):
        data[x,y] = numpy.sin(x/10) * numpy.sin(y/10) * numpy.sin(x*y/500)
        pass
    pass

plt.imshow(data, origin='lower')
plt.show()
```

## contour plot

use previous 2d data

```python
plt.contour(data)
plt.show()
```

## 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = numpy.arange(0,100,1)
y = numpy.arange(0,100,1)

xs, ys = numpy.meshgrid(x, y)
# z = calculate_R(xs, ys)
zs = numpy.sin(xs/10) * numpy.sin(ys/10) * numpy.sin(xs*ys/500)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, cmap='hot')
plt.show()
```

## xkcd style plot !

just for fun

```python
with plt.xkcd():

    X = numpy.linspace(0, 10, 200)
    Y = numpy.sin(X) * 1000

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.title("TOURISTS IN CORNWALL")

    plt.plot(X,Y)
    plt.show()
    
    pass
```

## one more thing .. plotting pandas dataframes

Several libraries try to make plotting easier than matplotlib. 

Plotting pandas dataframes is a good example.

```python
import pandas

# load csv into pandas dataframe
df = pandas.read_csv('my_data.csv', skipinitialspace=True, index_col='planet')

df
```

```python
df.plot(x='mass', y='brightness', kind='bar')
plt.show()
```

```python
df.plot(subplots=True, kind="bar")
plt.show()
```

```python
# use reference docs to make figure larger
# Reference http://pandas.pydata.org//pandas-docs//stable//generated/pandas.DataFrame.plot.html
```
